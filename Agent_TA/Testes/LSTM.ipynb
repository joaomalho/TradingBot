#### <font color='white'> __2.1.1 EMAs Crossover & Bollinger Bands & RSI__ </font> <a class="anchor" id="twooneone"></a>

[<font color='mediumpurple'>Back to Index </font>](#index)


```python
def add_metrics(data):
    '''
    This function will add the technical metric in the data, such as RSI, EMAs Crossover, SMA Bands 
    '''

    # ==== EMAs Crossover ==== #
    # EMA14 
    data['ema_13'] = pta.ema(data['close'], length=13)
    # EMA20    
    data['ema_48'] = pta.ema(data['close'], length=48)
    # EMA200
    data['ema_200'] = pta.ema(data['close'], length=200)

    def calculate_crossover(row):
        if row['ema_13'] > row['ema_48'] and row['ema_48'] > row['ema_200']:
            return 'Buy'
        elif row['ema_13'] < row['ema_48'] and row['ema_48'] < row['ema_200']:
            return 'Sell'
        else:
            return 'Flat'

    data['EMAs_Crossover_Signal'] = data.apply(calculate_crossover, axis=1)


    # ==== SMA Bands ==== #
    hour_data_bb = pta.bbands(data.close, length=SMA_PERIOD, std=STANDARD_DEVIATION_SMA)
    data = data.join(hour_data_bb)

    # Finding signal
    def calculate_sma_bands(row):
        if row['close'] <= row['BBL_14_2.0']:
            return 'Buy'
        elif row['close'] >= row['BBU_14_2.0']:
            return 'Sell'
        else:
            return 'Flat'

    data['SMA_Bands_Signal'] = data.apply(calculate_sma_bands, axis=1)


    # ==== RSI ==== #
    data['rsi'] = pta.rsi(data['close'], length=RSI_PERIOD)
    
    data['rsi_oversold_level'] = OVERSOLD_LEVEL
    data['rsi_overbought_level'] = OVERBOUGHT_LEVEL
    data['rsi_middle_level'] = MIDDLE_LEVEL

    def calculate_rsi(row):
        if row['rsi'] >= OVERBOUGHT_LEVEL:
            return 'Sell'
        elif row['rsi'] <= OVERSOLD_LEVEL:
            return 'Buy'
        else:
            return 'Flat'

    data['RSI_signal'] = data.apply(calculate_rsi, axis=1) 

    return data

data = add_metrics(data)
```

```python
# Candles Patterns

def candle_basic_signal(data):
    '''
    This function perform a for loop measure the candles types, bearish ou bullish and shadows size comparatively with body size.
    
    Conditions:
        * bullish = Close Price > Open Price
        * bearish = Close Price < Open Price
        * flat = Else
    '''

    data['candle_type'] = ''
    data['upper_shadow_size'] = ''
    data['lower_shadow_size'] = ''
    data['body_value_size'] = ''
    
    for index, row in data.iterrows():
        if row['close'] > row['open']:
            data.at[index, 'candle_type'] = 'bullish'
            
            upper_shadow = row['high'] - row['close']
            body = row['close'] - row['open']
            lower_shadow = row['open'] - row['low']
            
            data.at[index, 'upper_shadow_size'] = upper_shadow/body
            data.at[index, 'lower_shadow_size'] = lower_shadow/body
            data.at[index, 'body_value_size'] = body
            
        elif row['close'] < row['open']:
            data.at[index, 'candle_type'] = 'bearish'
            
            upper_shadow = row['high'] - row['open']
            body = row['open'] - row['close']
            lower_shadow = row['close'] - row['low']
            
            data.at[index, 'upper_shadow_size'] = upper_shadow/body
            data.at[index, 'lower_shadow_size'] = lower_shadow/body
            data.at[index, 'body_value_size'] = body
        else:
            data.at[index, 'candle_type'] = 'doji'
            
            upper_shadow = 0
            body = 0
            lower_shadow = 0
            
            data.at[index, 'upper_shadow_size'] = 0
            data.at[index, 'lower_shadow_size'] = 0
            data.at[index, 'body_value_size'] = 0
    
    return data

data = candle_basic_signal(data)
```

```python
def inverted_hammer(data):
    '''
    This function detects inverted hammer candlestick patterns and updates the respective columns.

    Fundamental:
        * The inverted hammer candlestick pattern indicates a bullish reversal or short-term downtrend reversal.
        An inverted hammer occurs after a prolonged sell-off when prices are near their lows for that period.

        * The inverted hammer is supposed to be a bullish reversal candlestick, but it really acts as a bearish continuation 65% of the time.
    '''

    data['inverted_hammer'] = (data['upper_shadow_size'] >= 2) & (data['lower_shadow_size'] < 1)

inverted_hammer(data)
```

#### <font color='white'> __2.1.5 Engolfing Candle__ </font> <a class="anchor" id="twoonefive"></a>

```python
def engolfing(data):
    '''
    This function detects if each candle is a bullish or bearish engulfing pattern.
    
    Fundamental:
        * In a Bullish Engulfing pattern, the body of the second (bullish) candle should completely engulf the 
        body of the first (bearish) candle.
        * In a Bearish Engulfing pattern, the body of the second (bearish) candle should completely engulf the 
        body of the first (bullish) candle.
        * Bullish engulfing has a reversal rate of 63%, while bearish engulfing has a reversal rate of 79%.
    '''
    data['bullish_engolfing'] = False
    data['bearish_engolfing'] = False
    
    for i in range(1, len(data)):
        current_day_candle = data.iloc[i]
        previous_day_candle = data.iloc[i - 1]
        
        bullish_engulfing = (
            current_day_candle['candle_type'] == 'bullish' and
            previous_day_candle['candle_type'] == 'bearish' and
            current_day_candle['body_value_size'] > previous_day_candle['body_value_size'] and
            current_day_candle['open'] < previous_day_candle['close'] and
            current_day_candle['close'] > previous_day_candle['open']
        )

        bearish_engulfing = (
            current_day_candle['candle_type'] == 'bearish' and
            previous_day_candle['candle_type'] == 'bullish' and
            current_day_candle['body_value_size'] > previous_day_candle['body_value_size'] and
            current_day_candle['open'] > previous_day_candle['close'] and
            current_day_candle['close'] < previous_day_candle['open']
        )

        data.at[i, 'bullish_engolfing'] = bullish_engulfing
        data.at[i, 'bearish_engolfing'] = bearish_engulfing

engolfing(data)
```

```python
def morning_star(data):
    '''
    A morning star is a three-candle pattern with the low point on the second candle. 
    However, the low point is only apparent after the close of the third candle.
    High volume on the third day is often seen as a confirmation of the pattern 
    (and a subsequent uptrend) regardless of other indicators.
    
    Fundamental:
        * First candle: The first candlestick is a tall bearish candle, in line with the ongoing downward price swing. It shows that the downswing is still intact.
        * Second candle: The second candlestick is a small-bodied candle, typically a doji or a spinning top. It gaps down from the first candlestick. It can be of 
        any color, but the most important thing is that it is small and gaps below the first, in anticipation of the continuation of the existing downswing. 
        Its typical small size suggests indecision in the market. If it is a doji, the pattern is called Morning Doji Star.
        * Third candle: The third candle is bullish and quite sizeable. It opens below the second candle and closes above the midpoint of the first candle. 
        * Bullish reversal 78% of the time.
        * Bearish reversal of the upward price trend, and testing reveals that it does 71% of the time.
    '''
    
    data['morning_star'] = 0
            
    third_candle = data.tail(1)
    second_candle = data.tail(2).head(1)
    first_candle = data.tail(3).head(1)

    data['morning_star'] = (first_candle.candle_type.item() == 'bearish' 
                                and second_candle.open.item() < first_candle.close.item() 
                                and second_candle.body_value_size.item() < first_candle.body_value_size.item()
                                and third_candle.candle_type.item() == 'bullish' 
                                and third_candle.open.item() < first_candle.close.item() 
                                and third_candle.close.item() > ((first_candle.open.item() - first_candle.close.item()) / 2)
                                )

morning_star(data)
```

```python
def get_sup_resis_fractal(data):
        '''
        This function measures the existence of Support and Resistences
        
        Fundamental:
            * The support and resistance (S&R) are specific price points on a chart expected to attract the maximum amount of either 
            buying or selling. 
            * Support - Is a price at which one can expect more buyers than sellers. 
            * Resistance - Is a price at which one can expect more sellers than buyers.
        
        Conditions:
            * Buy = Support
            * Sell = Resistnece
            * Flat = Else     
        '''
        
        # Hour View
        # To use fractal we have to regard the last 5 candles
        def isSupport_5candles(data,i):
            return (data['low'][i] < data['low'][i - 1] 
                    and data['low'][i] < data['low'][i + 1]
                    and data['low'][i + 1] < data['low'][i + 2]
                    and data['low'][i - 1] < data['low'][i - 2])
        def isResistance_5candles(data,i):
            return (data['high'][i] > data['high'][i - 1]
                    and data['high'][i] > data['high'][i + 1]
                    and data['high'][i + 1] > data['high'][i + 2]
                    and data['high'][i - 1] > data['high'][i - 2]) 

        levels_5candles = []
        for i in range(2,data.shape[0]-2):
            if isSupport_5candles(data,i):
                levels_5candles.append((i,data['low'][i],'support'))
            elif isResistance_5candles(data,i):
                levels_5candles.append((i,data['high'][i],'resistence'))

        # Check first positions, is note necessary to biggin in the 2nd place, although the measure take the first 5 candles, ALIGN ?????
        levels_5candles = pd.DataFrame(levels_5candles)
        levels_5candles.rename(columns={0:'sup_resis_level_position',
                                        1:'sup_resis_level_value',
                                        2:'sup_resis_level_type'},
                                inplace=True
                                )
        levels_5candles.set_index('sup_resis_level_position', inplace=True)
        
        data = data.join(levels_5candles, lsuffix=data.index, rsuffix=levels_5candles.index)
        
        return data
        
data = get_sup_resis_fractal(data)
```

Missing Values are always bad for modeling every task and cant exist in a dataframe in Data Egineering.

The correct approachs will depend of the type of missing value, if those are numeric can be replaced by the mean of the feacture values or even the mode but only if the percentage of existence be below 5 % if dont can be safely dropped from the data. If exist in a binary feacture (f.e. 1 - Exist/Occur, Nan - Dont Exist/Occur) should be replaced by 0, in categorical if should be replaced in a balanced way in order to dont overrate or downrate a class.

### __Data Engineering__

```python
def missing_values(data: pd.DataFrame, threshold: float = 0.02):
    """
    This functions will return a pandas dataframe with feature"s missing values count in absolute and relative frequency
    The threshold is the max % of missing values, for a given threshold this function will return if the features are 
    above or bellow the given threshold limit (True for above or False for bellow).
    
    Parameters:
    * Data - function expect to receive the data in a pandas Dataframe formate
    * Threshold - function expect to receive an input for threshold variable from 0 to 1 (0 = 0% and 1 = 100%)
    """

    # Count the number os missing values per each feacture
    na_count = data.isna().sum()
    # Filter data feactures that have missing values
    na_exist = na_count[na_count > 0]
    # Save an array with absolute amount of missing values
    na_abs_frq = na_exist.values
    # Save an array with relative % of missing values
    na_rel_frq = round(na_exist/len(data),2)

    # Creating a pandas DataFrame with columns "Feature", "NÂº of missings" and "% of missings" 
    missings = pd.DataFrame({"NÂº of missings": na_abs_frq, "% of missings": na_rel_frq})
    # Sort values by "NÂº of missings" in a descendent way 
    missings = missings.sort_values(by = "NÂº of missings", ascending = False)

    # If threshold given then the column "Above threshold" is created and filled with True if count (x) above threshold else False
    if threshold:
        missings["Above threshold"] = [True if x > threshold else False for x in missings["% of missings"]]
        
    return missings

missing_values(data)
```

```python
def check_zscore_outliers(data, col_name, threshold=3.5):
    '''Checks values that Z-score test consider as outliers
    
    Args: 
        data (pandas.core.frame.DataFrame): set of data 
        col_name (str): name of the column in data to check outliers
        threshold (float): the Z-score threshold for identifying outliers
        
    Returns: 
        outlier_positions (list): List of positions of identified outliers
    '''
    Z = np.abs(zscore(data[col_name]))  # Z-scores for the column
    positions = np.where(Z >= threshold)  # positions of identified outliers

    mini = data[col_name].min() 
    maxi = data[col_name].max() 
    print(f'The {col_name} feature ranges from {mini} to {maxi}\n\nOutlier values:\n')
    for row in positions[0]:
        print(f'Value from row {row}: ', data[col_name].iloc[row].round(2))
    
    return positions[0]

# Assuming 'data' is the DataFrame containing the data and you want to check for outliers in each column
# Select the features that you want to use for prediction (excluding 'time' and 'close')
tmp = data.drop(['time', 'close','inverted_hammer', 'hammer', 'morning_star',
                    'evening_star', 'upper_shadow_size', 'lower_shadow_size',
                    'body_value_size', 'bullish_engolfing', 'bearish_engolfing',
                    'morning_star', 'evening_star', 'candle_type_bearish',
                    'candle_type_bullish', 'candle_type_doji','sup_resis_level_type','comb_ema_bollinger'], axis=1).columns.tolist()

# Assuming 'data' is the DataFrame containing the data and you want to check for outliers in each column
all_positions = []  # List to store outlier positions for all columns
for col in tmp: 
    if data[col].dtype in [np.float64, np.int64]:
        positions = check_zscore_outliers(data, col)
        all_positions.extend(positions)
        print('\n')

print('If we eliminate outliers with the Z-Score Method it\'ll remain {} % of the initial data'.format(np.round((1 - len(set(all_positions)) / len(data)) * 100, 3)))
```

```python
def plot_normalization_scores(data):
    # Extract the feature columns from the DataFrame
    X = data.select_dtypes(include=[np.number])  # Select only numerical columns
    
    # Define the normalization functions
    def normalize_min_max_0_1(data):
        scaler = MinMaxScaler(feature_range=(0, 1))
        X_normalized = scaler.fit_transform(data)
        return pd.DataFrame(X_normalized, columns=data.columns)
    
    def normalize_min_max_m1_1(data):
        scaler = MinMaxScaler(feature_range=(-1, 1))
        X_normalized = scaler.fit_transform(data)
        return pd.DataFrame(X_normalized, columns=data.columns)
    
    def normalize_robust(data):
        scaler = RobustScaler()
        X_normalized = scaler.fit_transform(data)
        return pd.DataFrame(X_normalized, columns=data.columns)
    
    # Compute the scores for each normalization method
    scores = []
    for normalize_fn in [normalize_min_max_0_1, normalize_min_max_m1_1, normalize_robust]:
        X_normalized = normalize_fn(X)
        score = np.abs(X_normalized - X).mean().mean()
        scores.append(score)
    
    # Plot  
    plt.figure(figsize=(7, 5))
        
    # Plot the scores
    labels = ['MinMax [0,1]', 'MinMax [-1,1]', 'Robust']
    x = np.arange(len(labels))
    plt.bar(x, scores, color='LightSteelBlue', edgecolor='SteelBlue')
    plt.ylabel('Mean absolute error', fontsize=13, fontweight='bold')
    plt.xticks(x, labels, fontsize=13, fontweight='bold')
    
    # Increase the value to adjust the position of labels
    label_offset = 0.5
    
    for i, v in enumerate(scores):
        plt.text(x[i] - 0.2, v - label_offset, str(round(v, 3)), fontsize=17)
        
        
    plt.show()

# Call the function with your DataFrame
plot_normalization_scores(data)
```

Is the beahavour of Trend and Seasonal together, the comparison of the values at each point might be the exact same value in which case the residuals will be zero, in other cases the value is different so the Residuals show what value must be added to Trend and Seasonal in order to adjust the result and get the same value. Resisduals usually correspond to random erros, also termed _white noise_, which is the information that cannot model or predict since it is completly random. Is obtaind by subtrating trend and seasonal values from the corresponding observed values.

### <font color='white'> __3.1 Augmented Dickey-Fuller (ADF) test__ </font> <a class="anchor" id="threeone"></a>

[<font color='mediumpurple'>Back to Index </font>](#index)

* This test helps determine if a time series is stationary by testing the existence of a unit root, ___if a unit root is present___ then time series ___is not stationary___. The null hypothesis states that unit root is present, meaning that our time series is not stationary.

* If the test returns a ___p-value less than a significance level of 0.05___, then ___null hypothesis can be rejected___ meaning that there are ___no unit root___ and so the series ___is stationary___.

```python
def get_stationarity(data = data):

    stationary = 0
    ADF_result = adfuller(data['close'])
    print('1Âº Phase - Apply Dickey Fuller\n'.format(ADF_result[0]))
    print('ADF Statistic: {}'.format(ADF_result[0]))

    if ADF_result[1] > 0.05:
        
        print('p-value: {} -> Is Not Stationary\n'.format(ADF_result[1]))
        print('2Âº Phase - Due not being Stationary is necessary to apply first difference:\n')
        diff_close = np.diff(data['close'], n=1)
        diff_ADF_result = adfuller(diff_close)
        print('ADF Statistic: {}'.format(diff_ADF_result[0]))
        
        if diff_ADF_result[1] < 0.05:
            print('p-value: {}'.format(diff_ADF_result[1]))
            print('3Âº Phase - Once is Not Stationary it is necessary to apply second difference:\n')
            second_diff_close = np.diff(diff_close, n=1)
            second_diff_ADF_result = adfuller(second_diff_close)
            print('ADF Statistic: {}'.format(second_diff_ADF_result[0]))
    
            if second_diff_ADF_result[1] < 0.05:
                print('p-value: {} -> Is Stationary after difference then is a Random Walk\n'.format(round(second_diff_ADF_result[1])))
                stationary = 1
    
    else:
        pass
    return stationary, ADF_result

# Assuming you have a DataFrame named 'data'
result = get_stationarity(data)
print("Final Result:", result)
```

```python
from sklearn import metrics

def performance_metrics(y_Train, y_PredTrain, y_Test, y_PredTest):
    '''
    Creates Dataframe with the main benchmark metrics for Classfication models
    
    Args: 
        y_Train -      Series with target values used to train model
        y_PredTrain -  Series with predicted values from y_Train
        y_Test -       Series with test values used to validate model
        y_PredTest -   Series with predicted values from y_Test
        
    Returns: 
        pandas.core.frame.DataFrame
    '''
    # Columns
    measures_list = ['ACCURACY','PRECISION', 'RECALL','F1 SCORE','AUC']
    
    # Results for train data
    train_results = [
                round(metrics.accuracy_score(y_Train, y_PredTrain),3),  # Calculating accuracy
                round(metrics.precision_score(y_Train, y_PredTrain),3),  # Calculating precision
                round(metrics.recall_score(y_Train, y_PredTrain),3),  # Calculating recall
                round(metrics.f1_score(y_Train, y_PredTrain),3),  # Calculating f1 score
                round(metrics.roc_auc_score(y_Train, y_PredTrain),3)  # Calculating ROC area under the curve
                ]
    # Results for test data
    test_results = [
                round(metrics.accuracy_score(y_Test, y_PredTest),3),  # Calculating accuracy
                round(metrics.precision_score(y_Test, y_PredTest),3),  # Calculating precision
                round(metrics.recall_score(y_Test, y_PredTest),3),  # Calculating recall
                round(metrics.f1_score(y_Test, y_PredTest),3),  # Calculating f1 score
                round(metrics.roc_auc_score(y_Test, y_PredTest),3)  # Calculating ROC area under the curve
                ]
    
    results = pd.DataFrame({'Measure': measures_list, 'Train': train_results, 'Test': test_results})
    
    # Comparison column
    results['Difference'] = round(results['Train'] - results['Test'], 2) 
    
    return results
```

```python
def model_df(model_list, scaled_X_test, test_y):
    '''
    Creates Dataframe with the main benchmark metrics for Classification models
    
    Args: 
        model_list (list) - List with model instances
        scaled_X_test (dataframe) - Scaled test data
        test_y (dataframe) - Test target
        
    Returns: 
        pandas.core.frame.DataFrame
    '''
    # Creating lists to store metrics
    model_names = []
    f1_scores = []
    accuracy_scores = []
    precision_scores = []
    recall_scores = []
    auc_scores = []

    # Getting models' results
    for model in model_list:
        model_name = type(model).__name__  # Get the model name
        model_names.append(model_name)

        y_pred = model.predict(scaled_X_test)  # Scaled data prediction

        # Calculate benchmark metrics
        f1 = metrics.f1_score(test_y, y_pred, average='micro')
        accuracy = metrics.accuracy_score(test_y, y_pred)
        precision = metrics.precision_score(test_y, y_pred, average='micro')
        recall = metrics.recall_score(test_y, y_pred, average='micro')
        roc_auc = metrics.roc_auc_score(test_y, y_pred)

        # Append metrics to lists
        f1_scores.append(f1)
        accuracy_scores.append(accuracy)
        precision_scores.append(precision)
        recall_scores.append(recall)
        auc_scores.append(roc_auc)

    # Create DataFrame
    df = pd.DataFrame({
        'Model': model_names,
        'F1-Score': f1_scores,
        'Accuracy': accuracy_scores,
        'Precision': precision_scores,
        'Recall': recall_scores,
        'AUC': auc_scores
    })

    # Sort DataFrame by F1-Score
    df.sort_values(by='F1-Score', ascending=False, inplace=True)
    df.set_index('Model', inplace=True)

    return df
```
